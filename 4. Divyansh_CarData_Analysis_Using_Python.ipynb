```python
def replace_null_values(df):
    categorical_columns = df.select_dtypes(include='O').columns
    numerical_columns = df.select_dtypes(exclude='O').columns


    for i in categorical_columns:
        mode_value = df[i].mode()[0]
        df[i].fillna(mode_value, inplace=True)

  
    for j in numerical_columns:
        mean_value = df[j].mean()
        df[j].fillna(mean_value, inplace=True)

    return df
df=replace_null_values(df)
```

# Task 2: Pre process the data in the following columns as specified - {1} Remove "kg" and ensure the column is numerical and convert it to pounds by multiplying the number with 2.2. {2} All the entries in power column should be numerical {3} Torque must only contain numerical also

# Task 4: Identify at least three features that affect the mileage (using correlation) of cars from the following manufacturers: 'Renault,'Toyota','Honda'

```python
selected_manufacturers = ['Renault', 'Toyota', 'Honda']
filtered_data = df[df['Make'].isin(selected_manufacturers)]
shortlisted=filtered_data.loc[:,['City_Mileage_km_litre','Displacement', 'Cylinders', 'Valves_Per_Cylinder']]
shortlisted
```

## Interpretation: The visual shows that theres a negative relationship between city mileage and displacement, cylinders and valves per cylinder.
